```python
import pandas as pd
import numpy as np
import math
from scipy.stats import norm
import matplotlib.pyplot as plt
import dateutil
import calendar
```

```python
# basic functions
def calculateMean(inputList):
    sum = 0
    for element in inputList:
        sum += element
    return sum/len(inputList)
    
def calculateStandardDeviation(inputList):
#     1. Work out the Mean (the simple average of the numbers)
        mean = calculateMean(inputList)
    
#     2. Then for each number: subtract the Mean and square the result
        stdDevs = []
        for index, element in enumerate(inputList):
            result = (element - mean) ** (2)
            stdDevs.append(result)

#     3. Then work out the mean of those squared differences.
        stdDevMean = calculateMean(stdDevs)
    
#     4. Take the square root of that and we are done!
        return math.sqrt(stdDevMean)
```

```python
sample = norm.rvs(size=1000)

myMean = calculateMean(sample)
print("My mean: " + str(myMean))
numpyMean = sample.mean()
print("numpy mean: " + str(numpyMean))

mySD = calculateStandardDeviation(sample)
print("My SD: " + str(mySD))
numpySD = sample.std()
print("numpy SD: " + str(numpySD))

plt.hist(sample, bins='auto') 
plt.title("Histogram with 'auto' bins")
plt.show()
```

```python
# readying a data frame for the below graphs
df = pd.read_csv("/Users/willdunn/GitWillDone/Data_Analytics_and_Vis/epaData/2018-PM2.5.csv", usecols=[0,10], skiprows=[1,2])
df.columns.values[0]='time'
df['time'] = pd.to_datetime(df['time'])
```

```python
# readings from LN station over the course of a year
annualData = df['LN'][2:].astype(float).plot()
```

# Insights for mean monthly data:
The density of the particles in the air increases during the fall months.  While it does have a similar behavior during the other months of the year, it is less pronounced.

```python
# the mean pm2.5 level for each month using a bar chart.

dataByMonth = df.groupby(df.time.dt.month).mean().reset_index()
dataByMonth.columns.values[0]='months'
dataByMonth['months'] = dataByMonth['months'].apply(lambda x: calendar.month_abbr[x])

dataByMonth.plot(kind='bar',x='months',y='LN')
plt.show()
```

# Insights from hourly data
The mean pollution fluctuates mostly between 6 and 7 particles during each hour. A mean of means would give a snapshot of what could be exprected for an average hourly rate during the time period throughout the year.

```python
# Next, group the data by time of day (by hour), and plot the mean pollution level for each hour. 

dataByHour = df.groupby(df.time.dt.time).mean().reset_index()

dataByHour.plot(kind='bar',x='time',y='LN')

plt.show()
```

# Additional Insights from the Box and Whisker Plots
The box and whisker plots give a good deal more insight than the bar graphs (above).  
- We are given the median for each month and/or hourly encapsulation of data.  This tells us the halfway point of our data - meaning that everything above is the greater half of the data, and vice versa for below.  
- We are also given the range, shown from the top and bottom of T shapes
- The data is broken up into quartiles which allows us to glean insight into the frequency of where our data points are occurring.
- Finally, the outliers are illustrated with dots so we can determine which pieces of data might have been skewing our above plots.

```python
# The mean only gives us a very coarse view of the monthly/hourly data. Use Box and Whisker plots of the monthly and 
# hourly data groupings to provide a more complete view of the data.
dfMonthPlot = df.copy()
dfMonthPlot['time'] = dfMonthPlot['time'].dt.month
dfMonthPlot['time'] = dfMonthPlot['time'].apply(lambda x: calendar.month_abbr[x])
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
dfMonthPlot['time'] = pd.Categorical(dfMonthPlot['time'], categories=months, ordered=True)
dfMonthPlot.sort_values(by='time',inplace=True)
mPlot = dfMonthPlot.boxplot(by='time', figsize=(16,8), fontsize=(9))
mPlot.set(xlabel='Month', ylabel='PPM Level (LV)', ylim=(0, 85), title='Month Grouping')
plt.suptitle('') # used to remove a redundant titling of the graph

dfHourPlot = df.copy()
dfHourPlot['time'] = dfHourPlot['time'].dt.time
hPlot = dfHourPlot.boxplot(by='time', figsize=(16,8), fontsize=(9))
hPlot.set(xlabel='Hours', ylabel='PPM Level (LV)', ylim=(0, 85), title='Hourly Grouping')
plt.suptitle('')
plt.show()
```
