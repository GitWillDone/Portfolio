# Part 1: A/B Hypothesis testing
The null hypothesis for part 1 is that logo B is a more popular logo, in terms of ad clicks.

```python
import numpy as np
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
import locale
import matplotlib.pyplot as plt

counts = np.array([500, 550]) # A: 500; B: 550
nobs = np.array([1000, 1000]) # both sample sizes are 1000

zscore, pvalue = proportions_ztest(counts, nobs, alternative = 'two-sided')
print('zscore = {:.3f}, pvalue = {:.3f}'.format(zscore, pvalue)) # RESULTS: zscore = -2.239, pvalue = 0.025
```

# Findings for Part 1:
The null hypothesis is rejected a values lower than 2.5% therefor showing that it could not be rejected a 1% significance level, but a could at a 5% level.

# Part 2: Regression of real estate data

### Task 1: Import the data

```python
data = pd.read_csv("realEstate1.csv").append(pd.read_csv("realEstate2.csv"))
# data
```

###  Task 2: Clean the data

```python
# Only keep houses with List Price between 200,000 and 1,000,000 dollars. 
data = data[(data['LstPrice'] >= 200000) & (data['LstPrice'] <= 1000000)]

# Remove columns that don't contribute to the value of the house.
data.drop(data.columns.difference(['Acres', 'Deck', 'GaragCap', 'Latitude', 'Longitude', 'LstPrice', 'Patio', 'PkgSpacs', 'PropType', 'SoldPrice', 'Taxes', 'TotBed', 'TotBth', 'TotSqf', 'YearBlt']), 1, inplace=True)

# Check the datatypes and convert any numbers that were read as strings to numerical values. 
# In particular, convert 'TotSqf' to an integer and add a column titled Prop_Type_num that is 0 if i-th listing is a 
# condo or townhouse, or 1 if i-th listing is a single family house
data['TotSqf'] = pd.to_numeric(data['TotSqf'].astype(str).str.replace(',',''), errors='ignore')

data = pd.concat([data, pd.get_dummies(data['PropType'])], axis=1)
del data['Condo'], data['Townhouse']
data.rename(index=str, columns={'Single Family' : 'Prop_Type_Num'}, inplace=True)

for col in data.keys():
    if col != 'TotSqf':
        data[col] = pd.to_numeric(data[col], errors='ignore') 
        
# Remove the listings with erroneous 'Longitude' (one has Longitude = 0) and 'Taxes' values 
# (two have unreasonably large values).
data = data[(data['Taxes'] < 9999) & (data['Longitude'] != 0)]
data
```

# Task 5: 

```python
import statsmodels.formula.api as sm
soldPriceAndListPrice = sm.ols(formula="SoldPrice ~ LstPrice", data=data).fit()
soldPriceAndListPrice.summary()
```

```python
plt.scatter(x=data['LstPrice'],y=data['SoldPrice'],c='k',marker='*',label='LstPrice')
plt.plot(data['LstPrice'],soldPriceAndListPrice.predict(),'k',color='blue',linewidth=3)

plt.xlabel('List Price')
plt.ylabel('Sold Price')
plt.show()
```

### Interpretation and discussion

The intercept of the line is Î²Ì‚0 0=1.317e+04.  This means that without any listing price, the house will be sold 
for ~$13,170.  

The slope of the line is Î²Ì‚1=0.9441, which means that hte model predicts that for every additional 1 dollar added to the list price, an additional $0.94 is added to the Sold Price.  This is a great fit because the R**2 value is so high
.990.

### Interpretation and discussion
For each additional square foot in the house, we can expect the value to increase by 99.78 dollars. For each additional garage, the value should rise by 41,480 dollars, but has a standard deviation of 28,200 dollars thus being a high difference for how much effect the garage capacity could actually have.  Latitude and Longitude exhibit strange results (2.902e6 and 2.852e6 respectively) with a massive standar deviation on both, that look extremely high for each numeric increase.  Even jupyter-notebook warns that the condition number is large and strong multicollinearity could be occurring.
