```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D
from keras.optimizers import SGD
from keras.callbacks import TensorBoard
from sklearn.model_selection import train_test_split, cross_val_predict, cross_val_score, KFold
import matplotlib.pyplot as plt
```

# Step 1: Data acquisition + cleanup

Pick a font to work with, from the dataset given, and load it into a pandas dataframe.

The data is unfortunately in a really, really horrible format for us. We're going to throw away basically everything except for the m_label field (the unicode codepoint for the character, which is the ascii value for basic alphanumeric characters) and the pixel values which are scattered across 400 columns labeled rxcy (where x and y are the row and column numbers that range from 0 to 19).

Write a function that transforms a dataframe and returns 2 numpy arrays: Xs which is a #samples x 20 x 20 array containing the pixel values, and Ys which is a #samples x 1 array containing the ascii vales for each character. You should divide everything in the Xs array by 255 to scale the values from 0-1.

```python
def readData(file):    
    df = pd.read_csv(file)
    y = np.zeros(len(df[['m_label']])) #because we only want m_label
    for i in range(len(df[['m_label']])):
        y[i] = df['m_label'][i]
        
    header = list(df) #TODO Why am I getting a header
    header = header[12:] # header of rxcy
#     print(header)
    X = np.zeros((len(y), 20, 20)) #making an empty array of samples * 20*20 matrix
    for i in range(len(y)):
        for j in range(20):
            for k in range(20):
                X[i][j][k] = df[header[j * 20 + k]][i] / 255 # scales the values from 0-1.
    return X, y
    
```

```python
X, y = readData('fonts/AGENCY.csv')
```

# Step 2: Preparing data for a neural net

Our data isn't quite ready for the neural net. We want to convert our labels (the y values) in to a "one-hot" representation. We'll do this in 2 steps:

First, the ascii values are fairly arbitrary, and we're skipping a bunch of ascii characters (character 0, the null character, for example). We want to assign each of the characters in our set an integer from 0-#(unique-characters). See the code in the rnn.ipynb for computing the char_to_ix and ix_to_char arrays, and write a function that computes those maps given a y vector.

Our network is going to output the probability of an input being one of the characters. For our training data, the probability of the character we know is 1, and the probability of it being anythign else is 0. Convert your y labels into their "1-hot" representation -- this should be a nSamples x nUniqueCharacters array with a single 1 entry in each row.

```python
# DO NOT RUN THIS MORE THAN ONCE WITHOUT RESETTING THE KERNEL!!! YOU ARE ALTERING Y AND WILL THROW AN ERROR
def oneHotRepresentation(y):
    data = y
    chars = list(set(data)) # remove duplicates to get amount of characters in the vector
    print ('data has %d characters, %d unique.' % (len(data), len(chars)))
    
    #Copied from the rnn.ipynb:
    # what is our "alphabet" ?  We're going to number the characters from 0 to numUniqueCharacters
    # and store forward and reverse mappings (what index is this char, what char is at this index(ix)?)
    char_to_ix = { ch:i for i,ch in enumerate(chars) }
    ix_to_char = { i:ch for i,ch in enumerate(chars) }
    
    oneHotEncoded = list()
    for value in data:
        letter = [0 for _ in chars] # if this doesn't work, use range(len(chars)); char is anything else 0%
        letter[char_to_ix[value]] = 1 # the probability of the char we know
        oneHotEncoded.append(letter)
    return oneHotEncoded, chars
    
y, chars = oneHotRepresentation(y)
```

# Step 3: Build a keras network

To start with, we're going to use a model very similar to the MNIST CNN we used with tensorflow (see the CNN notes). It will consist of:

    - a Convolution2D layer with relu activations
    - a max pooling layer
    - another convolution layer
    - another max pooling layer
    - a Flatten layer (this just reshapes the data)
    - a dropout layer (for improved training performance + overfitting avoidance)
    - a dense layer with relu activation
    - a dense layer with softmax activation (our output layer)

You'll need to specify a few parameters for each of these layers. Important parameters are the input_size for the first layer, and the number of units (which should be the number of unique characters) for the output layer.

Compile and train your network as shown in the Keras doc. You'll probably have to use the np.reshape function on your data to make Keras happy. I reshaped my X values like np.reshape(Xs, (-1, 20, 20, 1)) to get them in the right format.

```python
def makeModel(chars):
    model = Sequential() # FFN
    model.add(Conv2D(32, (4, 4), activation='relu', input_shape=(20, 20, 1))) #worked without -1. remove if issues
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(64, (4, 4)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dropout(0.4))
    model.add(Dense(10, activation='relu'))  # 64 works; but there are 251 unique chars
    model.add(Dense(len(chars), activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  optimizer='rmsprop',
                  metrics=['accuracy'])
    return model

model = makeModel(chars)
```

```python
#start by getting the reshaped sets
def getSets(X, y):
    X = np.reshape(X, (-1, 20, 20, 1)) 
    # 1 is random_state because it is either the number or not; .33 is what the sklearn website recommends for t_size
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1) # .2 works too t_size
    return X_train, X_test, y_train, y_test 

#train the model
def trainModel(model, X_train, y_train):
    #fit/train data
    model.fit(np.array(X_train), np.array(y_train), epochs=50) #batch-size default is 32  
    scores = model.evaluate(np.array(X_train), np.array(y_train))
    print("\nTraining score:")
    print(model.metrics_names)
    print(scores)
    
#run the test data through the model
def testModel(model, X_test, y_test):
    scores = model.evaluate(np.array(X_test), np.array(y_test))
    print("\nTesting score:")
    print(model.metrics_names)
    print(scores)

    
#execute the functions, in order
X_train, X_test, y_train, y_test = getSets(X,y)
trainModel(model, X_train, y_train)
testModel(model, X_test, y_test)

```

```python
def makeBetterModel(chars):
    model = Sequential()
    model.add(Conv2D(32, (4, 4), activation='relu', input_shape=(20, 20, 1)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(64, (4, 4)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dropout(0.25))
    model.add(Dense(1024, activation='relu')) # this needs to be at least 251 to outperform the last model
    model.add(Dropout(0.25))
    model.add(Dense(251, activation='relu'))
    model.add(Dense(len(chars), activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  optimizer='rmsprop',
                  metrics=['accuracy'])
    return model
    
betterModel = makeBetterModel(chars)
trainModel(betterModel, X_train, y_train)
testModel(betterModel, X_test, y_test)
```

The better model performs better in training (88% vs. 61%) and in testing (41% vs. 18%).

```python
def newEncode(data, chars):
    print ('Data has %d characters, %d unique.\n' % (len(data), len(chars)))
    char_to_ix = { ch:i for i,ch in enumerate(chars) }
    ix_to_char = { i:ch for i,ch in enumerate(chars) }
    oneHotEncoded = list()
    for value in data:
        letter = [0 for _ in range(len(chars))]
        letter[char_to_ix[value]] = 1
        oneHotEncoded.append(letter)
    return oneHotEncoded

X1, y1 = readData('fonts/AGENCY.csv')
X2, y2 = readData('fonts/English.csv') #English used because it has the same number of unique chars

y_combine = np.append(y1, y2)
chars = list(set(y_combine))

y1 = newEncode(y1, chars)
betterModel = makeBetterModel(chars)
X1 = np.reshape(X1, (-1, 20, 20, 1))
trainModel(betterModel, X1, y1)

y2 = newEncode(y2, chars)
X2 = np.reshape(X2, (-1, 20, 20, 1))
testModel(betterModel, X2, y2)
```

```python
X1, y1 = readData('fonts/AGENCY.csv')
X2, y2 = readData('fonts/English.csv')
X3, y3 = readData('fonts/BELL.csv') # Also selected for the number of similar characters

y_combine = np.append(y1, y2)
y_combine = np.append(y_combine, y3)

chars = list(set(y_combine))
betterModel = makeBetterModel(chars)

X_train = np.append(X1, X2)
y_train = np.append(y1, y2)
y_train = newEncode(y_train, chars)

X_train = np.reshape(X_train, (-1, 20, 20, 1))
trainModel(betterModel, X_train, y_train)

y_test = newEncode(y3, chars)
X_test = np.reshape(X3, (-1, 20, 20, 1))
testModel(betterModel, X_test, y_test)
```

#### We can see that training on multiple similar sets (in their unique characters) slightly improves the accuracy of our training ( 91% vs 88%) and vastly improves accuracy in testing ( 40% vs 22%) sets.

```python
y_pred = betterModel.predict(X_test, batch_size = 32, verbose = 0, steps=None)
misclass = np.zeros(100)
count = 0
for i in range(len(y_pred)):
    if (y_pred[i] != y_test[i]).all():
        misclass[count] = i
    count += 1
    if count == 100:
        break
        
plt.figure(figsize= (10, 10))    
for i in misclass:
    ii = int(i)
    plt.subplot(10, 10, ii+1)
    plt.imshow(np.reshape(X3[ii],(20, 20)), cmap='Greys',interpolation='nearest')
    plt.axis('off')
plt.show()
```

Above, we can see that many of the misclassed characters are fairly complex.  It is interesting to note that many misclasses involve accent marks.  It is possible that the model assumed these were all the same in their class (like the variations of 'o' in the bottom row) or found them to be similar to other vowels not in their class (a o and u with an umlaut might have been indistinguishable because the umlaut was what the model ended up prioritizing).

# Step 5: Denoising

Next we'll build and train a neural network (an autoencoder) for a different task: denoising images.

First, create a dataset which contains images from one or more fonts + noise (see the autoencoder notebook for an example of this).

Next, create and train a convolutional autoencoder to denoise these images.

Create a plot showing the noisy and denoised versions of some of the test inputs to verify that your denoiser had the desired effect.

```python
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D
from keras.models import Model
from keras import backend as K

input_img = Input(shape=(20, 20, 1))  # adapt this if using `channels_first` image data format

# very similar to what we did last week
x = Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
conv_encoded = MaxPooling2D((2, 2), padding='same')(x)
# conv_encoded.get_shape()

# # at this point the representation is (3, 3, 8) i.e. 72-dimensional
x = Conv2D(8, (3, 3), activation='relu', padding='same')(conv_encoded)
x = UpSampling2D((2, 2))(x) # max pooling -> upsampling 2D
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(16, (3, 3), activation='relu')(x)
x = UpSampling2D((2, 2))(x)
conv_decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

conv_autoencoder = Model(input_img, conv_decoded)
conv_autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
```

```python
# import our dataset - discarding labels later because we're autoencoding f(x) = x
# x, y = readData('fonts/Bell.csv') # smaller data set to ensure basic functionality first
x, y = readData('fonts/OCRA.csv') # a larger data set (font with more than 1000 ideally) - OCRA.csv mayhap
x_train, x_test, _, _ = getSets(x,y) 

#the data is already scaled in getSets(), so we'll just go ahead and start our reshaping
x_train_shaped = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test_shaped = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))

x_train_2d = np.reshape(x_train_shaped, (x_train_shaped.shape[0], 20, 20 , 1)) #reshapes into size 640x400
x_test_2d = np.reshape(x_test_shaped, (x_test_shaped.shape[0], 20, 20 , 1))

x_train_2d.shape
#x_test_2d.shape
```

```python
conv_autoencoder.fit(x_train_2d, x_train_2d,
                epochs=50,
                batch_size=128,
                shuffle=True,
                validation_data=(x_test_2d, x_test_2d),
                callbacks=[TensorBoard(log_dir='/tmp/conv_autoencoder')])
#TODO should the samples be larger (original mnist is 60000 and 10000)
```

```python
# encode and decode some digits
# note that we take them from the *test* set
autoencoded_imgs_cnn = conv_autoencoder.predict(x_train_2d)

n = 10  # how many digits we will display
plt.figure(figsize=(20, 6))
for i in range(n):
    # display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_train_shaped[i].reshape(20, 20), cmap="Greys", interpolation=None)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(autoencoded_imgs_cnn[i].reshape(20, 20), cmap="Greys", interpolation=None)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    # display reconstruction from the first
plt.show()
```

## Denoise

```python
noise_factor = 0.1 # 0.0 will show the negative of each font shape 
x_train_noisy = x_train_2d + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train_2d.shape) 
x_test_noisy = x_test_2d + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test_2d.shape) 

x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)
```

```python
input_img = Input(shape=(20, 20, 1))  

x = Conv2D(32, (3, 3), activation='relu', padding='same', name = 'layer1')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
denoiser_encoded = MaxPooling2D((2, 2), padding='same')(x)

# at this point the representation is (7, 7, 32)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(denoiser_encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
denoiser_decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

denoiser = Model(input_img, denoiser_decoded)
denoiser.compile(optimizer='adadelta', loss='binary_crossentropy')
```

```python


denoiser.fit(x_train_noisy, x_train_2d,
                epochs=50,
                batch_size=128,
                shuffle=True,
                validation_data=(x_test_noisy, x_test_2d),
                callbacks=[TensorBoard(log_dir='/tmp/denoiser', histogram_freq=0, write_graph=False)])

```

```python
# encode and decode some digits
# note that we take them from the *test* set
denoised_images = denoiser.predict(x_test_noisy)

n = 10  # how many digits we will display
plt.figure(figsize=(30, 6))
for i in range(n):
    # display original
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(x_test_2d[i].reshape(20, 20), cmap="Greys", interpolation=None)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
    # noisy version
    ax = plt.subplot(3, n, n+ i + 1)
    plt.imshow(x_test_noisy[i].reshape(20, 20), cmap="Greys", interpolation=None)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display denoised version
    ax = plt.subplot(3, n, i + 1 + 2*n)
    plt.imshow(denoised_images[i].reshape(20, 20), cmap="Greys", interpolation=None)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
plt.show()
```
